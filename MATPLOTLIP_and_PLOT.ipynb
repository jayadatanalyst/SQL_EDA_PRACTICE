```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

<h4> LINE Chart - showing sales trend over months</h4>

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [1000, 1500, 1300, 1800]
```

```python
plt.plot(months,sales)
plt.title("Monthly Sales Trends")
plt.xlabel('months')
plt.ylabel('sales')
plt.show()
```

```python

data = {
'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
'sales': [1000,1200, 1500, 1300,1600,1800]
}
df = pd.DataFrame(data)

plt.plot(df['month'], df['sales'], marker ='o', color ='green')        
plt.title("Monthly Sales Trends")
plt.xlabel('month')
plt.ylabel('sales')
plt.grid(True)
plt.show()
```

<h4>Scatter Plot - Relationship between age and amount spent</h4>

```python
customer_age = [22,25,30,35,40]
order_amount = [200,250,400,350,400]

plt.scatter(customer_age, order_amount)
plt.title("customer age vs order amount")
plt.xlabel('customer_age')
plt.ylabel('order_amount')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

order_amounts = [100,150,200,250,300,350,150,200,180,400,350]

plt.hist(order_amounts, bins=5, color='skyblue', edgecolor ='black')
plt.title('Distribution of order amounts')
plt.xlabel('order amount')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.kdeplot(order_amounts, shade=True, color='purple')
plt.title('Distribution of order amounts')
plt.xlabel('order amount')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.kdeplot(order_amounts, fill=True, color='purple')
plt.title('Distribution of order amounts')
plt.xlabel('order amount')
plt.ylabel('Frequency')
plt.show()
```

```python
orders = [20,2,5,3,5,4,4]
customer_age = [22,25,30,35,40,28,33]
spending = [200,250,400,350,400, 280, 320]

plt.scatter(customer_age, spending, s=np.array(orders)*50, alpha=0.5)
plt.title('age vs spending (Bubble = orders)')
plt.xlabel('customer_age')
plt.ylabel('spending')
plt.show()
```

```python
from sklearn.cluster import KMeans
import seaborn as sns

data = {

    'CustomerID': [1,2,3,4,5],
    'Spending': [200,300,400,800,900],
    'Frequncy': [2,3,1,5,6]
}

df = pd.DataFrame(data)

kmeans = KMeans(n_clusters = 2)
df['Cluster'] = kmeans.fit_predict(df[['Spending', 'Frequncy']])

sns.scatterplot(x='Spending', y='Frequncy', hue='Cluster', data=df)
plt.title('customers clusters')
plt.show()
    
```
